```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def get_seq(s1,w):
    seq = [s1]
    for i in range(0,49):
        seq.append(w*seq[i]-w*(seq[i]**2))
    return seq

plt.plot(get_seq(0.5,1),marker='o',color = 'k')

plt.plot(get_seq(0.5,3),marker='o',color = 'b')

plt.plot(get_seq(0.0001,4),marker='o',color = 'r')

# pretty up graph
plt.xlabel('step')
plt.ylabel('value')
plt.legend(['s=0.5 w=1','s=0.5 w=3','s=0.0001 w=4'],loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
